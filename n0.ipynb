```python
import time
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def f(n):
    x=1
    start=time.time()
    for i in range(1,n+1):
        for j in range(1,n+1):
             x=x+1
    end=time.time()
    return end-start
values=np.arange(1,1001)
exe_times=[f(n) for n in values]
plt.plot(values,exe_times,label='Execution Time',color='blue',alpha=0.5)
def poly_curve(n,a,b,c):
    return a*np.power(n,2)+b*n+c
popt,pcov=curve_fit(poly_curve,values,exe_times)
plt.plot(values,poly_curve(values,*popt),color='red',label='Polynomial Fit')
n_0=np.argmin(np.abs(exe_times-poly_curve(values,*popt)))
plt.scatter(values[n_0],exe_times[n_0],color='black',label='n_0')
plt.xlabel('n')
plt.ylabel('Time (s)')
plt.title('Execution Time vs n')
plt.legend()
plt.grid(True)
plt.show()
print("Approximate value of n_0:",values[n_0])
```
