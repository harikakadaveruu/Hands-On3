```python
import time
import matplotlib.pyplot as plt
def f(n):
    x=1
    for i in range(1,n+1):
        for j in range(1,n+1):
            x=x+1
values=list(range(1,1001))
exe_times=[]
for n in values:
    start=time.time()
    f(n)
    end=time.time()
    exe_times.append(end-start)
plt.plot(values,exe_times,label='Execution Time')
plt.xlabel('n')
plt.ylabel('Time (s)')
plt.title('Execution Time vs n')
plt.legend()
plt.show()
```
